```python
import pandas as pd
import numpy as np
df = pd.read_csv("dropout-ratio-2012-2015.csv")
df.head(5)
```

```python
df.columns
```

```python
df.isna().any()
```

```python
df["State_UT"].unique()
```

```python
West_Bengal = df[ df["State_UT" ] == "West Bengal"]
West_Bengal
```

```python
df["Primary_Boys"].values
#มีติดNR - No Report
```

```python
df["Primary_Boys"].replace({"NR":0},inplace = True)
df["Primary_Boys"].values
```

```python
df.replace({"NR":0,"Uppe_r_Primary":0},inplace = True)
df[['Primary_Boys', 'Primary_Girls', 'Primary_Total',
       'Upper Primary_Boys', 'Upper Primary_Girls', 'Upper Primary_Total',
       'Secondary _Boys', 'Secondary _Girls', 'Secondary _Total',
       'HrSecondary_Boys', 'HrSecondary_Girls', 'HrSecondary_Total']] = df[['Primary_Boys', 'Primary_Girls', 'Primary_Total',
       'Upper Primary_Boys', 'Upper Primary_Girls', 'Upper Primary_Total',
       'Secondary _Boys', 'Secondary _Girls', 'Secondary _Total',
       'HrSecondary_Boys', 'HrSecondary_Girls', 'HrSecondary_Total']].astype(np.float32)
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (75,10))
sns.countplot(data=df, x='State_UT')
```

```python
df.groupby("year").mean()[["Primary_Boys","Primary_Girls"]].plot(kind = "bar")
plt.title("AVG dropout ratio by year in India")
plt.ylabel("Ratio (%)")
```

```python
plt.figure(figsize = (8,5))
bar_width = 0.2

x = np.arange(len(West_Bengal["year"]))

plt.bar(x ,West_Bengal["Primary_Girls"].astype(np.float32), bar_width)
plt.bar(x+bar_width,West_Bengal["Primary_Boys"].astype(np.float32), bar_width)

plt.xlabel("Year")
plt.ylabel("Dropout Ratio (%)")
plt.title("Dropout Ratio by Year in West Bengal Primary school")

for x,y in enumerate(West_Bengal["Primary_Girls"]) :
  plt.text(x-0.06,float(y)+0.05,y)
for x,y in enumerate(West_Bengal["Primary_Boys"]) :
  plt.text(x-0.06+bar_width,float(y)+0.05,y)

plt.xticks(np.arange(len(West_Bengal["year"])) + bar_width/2, West_Bengal["year"])
plt.legend(["Primary_Girls","Primary_Boys"])

plt.show()
```

```python
plt.figure(figsize = (28,15))
title = ['Primary_Boys', 'Primary_Girls', 'Upper Primary_Boys', 'Upper Primary_Girls','Secondary _Boys', 'Secondary _Girls',
       'HrSecondary_Boys', 'HrSecondary_Girls']

df[title].hist(figsize = (15,10),grid = False)
```

```python
plt.figure(figsize = (28,15))
#เลือกพิจารณาcentral part ของdata distribution (90%)

cenDf = df[title].apply(lambda x: x.clip(lower = x.quantile(0.05), upper = x.quantile(0.95)))

sns.boxplot(data = cenDf,orient = "h")
plt.xlabel("Ratio (%)")

plt.show()
```
